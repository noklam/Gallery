```python
import altair as alt
from vega_datasets import data
alt.data_transformers.enable('json')
iris = data.iris()

chart = alt.Chart(iris).mark_point().encode(
    x='petalLength',
    y='petalWidth',
    color='species'
)

chart
```

```python
chart.save('test.html')
```
